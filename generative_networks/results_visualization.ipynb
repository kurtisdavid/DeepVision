```python
for k in range(0,100,5):
    e0 = np.load('./results/anime_epoch_' + str(k) + '.npy')
    fig,axes = plt.subplots(3,4)
    for i in range(3):
        for j in range(4):
            axes[i,j].imshow(vf(e0[i*4 + j]).clip(0,255).astype('uint8'))
    fig.set_size_inches(15,10)
    fig.suptitle('Epoch ' + str(k), fontsize=18, fontweight='bold')
    fig.savefig('./results/anime_faces_epoch_ ' + str(k) + '.png')
    del fig
```
